# Trying to get the data from the APIs

Libraries imported

```python
from urllib.request import urlopen
import json
import pandas as pd
import geopandas as gpd
```

- ### Unitated States API

Opening the url as a json file

```python
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2014-01-01&endtime=2014-01-02'
json_obj = urlopen(url)
data = json.load(json_obj)
```

```python
data
```

```python
data['features']
```

```python
df = pd.json_normalize(data)
```

```python
df.info()
```

```python
df = pd.json_normalize(data, record_path=['features'])
```

```python
df = pd.json_normalize(data, record_path=['features'], meta=['properties.types'])
```

```python
df = pd.json_normalize(data, record_path=['bbox']) # this record path doesn't seem to give any valuable information
```

### Getting the data with location parameters, refering to the rectangle intersection of coordinates for the Unitated States

```python
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2014-01-01&endtime=2014-01-02&minlatitude=24.911795455444313&minlongitude=-124.65301770531302&maxlatitude=49.02924641916901&maxlongitude=-66.95789388605114'
json_obj = urlopen(url)
data = json.load(json_obj)
```

```python
df = pd.json_normalize(data, record_path=['features'])
```

```python
df
```

```python
df['geometry.type'].unique()
```

```python
df = pd.json_normalize(data, record_path=['features'], meta=['features', 'properties', 'types', 'nearby-cities'])#, 'nearby-cities'])
```
